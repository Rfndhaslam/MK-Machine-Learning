Dataset berisi kolom-kolom berikut:
1. jenis kelamin: Jenis kelamin siswa.
2. ras/etnis: Pengenal kelompok etnis.
3. tingkat pendidikan orang tua: Tingkat pendidikan orang tua siswa.
4. makan siang: Jenis makan siang (standar atau gratis/berkurang).
5. kursus persiapan tes: Status kursus persiapan tes (selesai atau tidak).
6. nilai matematika: Nilai dalam matematika.
7. skor membaca: Skor dalam membaca.
8. skor menulis: Skor dalam menulis.

```python
import pandas as pd

# Load the dataset provided by the user
file_path = 'StudentsPerformance.csv'
df = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure and contents
df.head()
```

1. Nilai Matematika: Distribusinya cukup merata, dengan puncaknya sekitar 65-75, yang menunjukkan bahwa sebagian besar siswa mendapat nilai dalam kisaran ini. Ada penurunan bertahap menuju nilai yang lebih tinggi.
2. Nilai Membaca: Distribusi ini sedikit bergeser ke arah nilai yang lebih tinggi dibandingkan dengan matematika. Banyak siswa mendapat nilai antara 70 dan 85, yang menunjukkan kinerja yang secara umum lebih baik dalam membaca daripada matematika.
3. Nilai Menulis: Sama halnya dengan membaca, nilai menulis juga menunjukkan puncak pada kisaran yang lebih tinggi, sekitar 65-85, yang menunjukkan bahwa siswa cenderung memiliki kinerja yang sedikit lebih baik dalam menulis dibandingkan dengan matematika.





```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
plt.style.use('default')

# Visualizing the distribution of scores in math, reading, and writing
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Score Distributions in Math, Reading, and Writing")

# Plotting math score distribution
sns.histplot(df['math score'], kde=True, ax=axes[0])
axes[0].set_title("Math Score Distribution")
axes[0].set_xlabel("Math Score")

# Plotting reading score distribution
sns.histplot(df['reading score'], kde=True, ax=axes[1])
axes[1].set_title("Reading Score Distribution")
axes[1].set_xlabel("Reading Score")

# Plotting writing score distribution
sns.histplot(df['writing score'], kde=True, ax=axes[2])
axes[2].set_title("Writing Score Distribution")
axes[2].set_xlabel("Writing Score")

plt.show()
```

* Korelasi Positif yang Kuat: Terdapat korelasi positif yang tinggi antara nilai membaca dan menulis (0,95). Hal ini menunjukkan bahwa siswa yang memiliki nilai bagus dalam membaca juga cenderung memiliki nilai bagus dalam menulis.
* Korelasi Positif Sedang: Baik nilai matematika dan membaca, maupun nilai matematika dan menulis, menunjukkan korelasi positif yang sedang (sekitar 0,6). Hal ini menunjukkan bahwa siswa dengan nilai matematika yang lebih tinggi umumnya cenderung memiliki nilai yang lebih baik dalam membaca dan menulis, meskipun hubungannya tidak sekuat hubungan antara membaca dan menulis.




```python
# Visualizing the correlation between math, reading, and writing scores
plt.figure(figsize=(8, 6))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix for Math, Reading, and Writing Scores")
plt.show()
```

* Siswa Perempuan: Mereka cenderung mendapat nilai lebih tinggi dalam membaca dan menulis dibandingkan dengan siswa laki-laki.
* Siswa Laki-laki: Mereka cenderung memiliki nilai yang sedikit lebih baik dalam matematika, meskipun perbedaannya minimal.

```python
# Visualizing average scores by gender
gender_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Plotting the scores by gender
gender_scores.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.legend(title="Subjects")
plt.show()
```

* Higher Education, Higher Scores: Siswa yang orang tuanya memiliki tingkat pendidikan yang lebih tinggi (seperti gelar master atau sarjana) cenderung mendapat nilai lebih tinggi di semua mata pelajaran, yang mengindikasikan hubungan positif antara pendidikan orang tua dan kinerja siswa.
* Lower Scores for Lower Parental Education: Siswa yang orang tuanya menyelesaikan sekolah menengah atas atau di bawahnya cenderung memiliki nilai rata-rata yang lebih rendah, menunjukkan bahwa latar belakang pendidikan orang tua dapat mempengaruhi kinerja akademik.

```python
# Visualizing average scores by parental level of education
parental_education_scores = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

# Plotting the scores by parental level of education
parental_education_scores.plot(kind='bar', figsize=(12, 6))
plt.title("Average Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xlabel("Parental Level of Education")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Subjects")
plt.show()
```

* Standard Lunch: Siswa dengan jenis makan siang standar cenderung mendapat nilai lebih tinggi di semua mata pelajaran dibandingkan dengan siswa yang mendapat makan siang gratis atau dikurangi, yang mungkin mengindikasikan bahwa faktor ekonomi yang terkait dengan jenis makan siang berdampak pada kinerja akademik.
* Free/Reduced Lunch: Siswa dalam kategori ini memiliki nilai rata-rata yang lebih rendah dalam matematika, membaca, dan menulis, yang menunjukkan bahwa dukungan ekonomi mungkin terkait dengan perbedaan hasil akademik.

```python
# Visualizing average scores by lunch type
lunch_scores = df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Plotting the scores by lunch type
lunch_scores.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Lunch Type")
plt.ylabel("Average Score")
plt.xlabel("Lunch Type")
plt.xticks(rotation=0)
plt.legend(title="Subjects")
plt.show()
```
